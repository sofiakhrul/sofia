Рассмотрим плотность распределения экспериментальных данных, полученных при измерении масс грузиков 

```python
%matplotlib inline

import pandas

df = pandas.read_csv("experiments.csv")

df['mass'].plot(kind='bar')

print(df.mass.describe())

df.plot.kde()
```

Проверим данные с помощью критерия Колмогорова-Смирнова

```python
from scipy import stats

d = df['mass']

print(stats.kstest(d , 'norm', (df.mean(), df.std())))
```

Ks-статистика маленькая, но Р-значение тоже очень маленькое. Видимо, не такая уж и нормальная выборка получилась - жаль(
